```python
# open and read txt file
with open('text.txt', 'r') as inp:
    text = inp.read()

# get number of all symbols
n = len(text)

# create list to store the frequency of each symbol
freqs = {}

# cycle to count frequencies
for i in text:
    #every time when we need the symbol it increments
    if i in freqs:
        freqs[i] += 1
    # if no more such symbol meets frequency remains 1
    else:
        freqs[i] = 1

# result shows the symbol and its probability with frequency
print(f'Nuber of all symbols: {n}')
print("Symbol  Probability           Frequency")
for i in freqs:
    print(f' {i}      {freqs[i] / n}   {freqs[i]}')
```
